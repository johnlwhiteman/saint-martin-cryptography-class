```python
%matplotlib inline
import numpy as np
import importlib
mpl_toolkits = importlib.import_module('mpl_toolkits')
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def show(prNums, asInts=False):
    if asInts:
        with np.printoptions(threshold=np.inf, suppress=True):
            print(prNums.astype(int))
    else:
        fmt = "{:.3f}".format
        with np.printoptions(threshold=np.inf, suppress=True, formatter={'float_kind':fmt}):
            print(prNums / m)

def plot(prNums):
    plt.figure(1)
    plt.hist(prNums, bins = 10)
    plt.title('LCG', fontsize = 19)
    plt.xlabel('Bins', fontsize = 16)
```

```python
def lcg(x0, cnt, a, c, m, scale=False):
    prNums = np.zeros(cnt)
    for i in range(0, cnt):
        x1 = (a * x0 + c) % m
        prNums[i] = x1
        x0 = x1
    if scale:
        return prNums / m  # (0,1)
    return prNums



a = 75
c = 74
m = 65537
seed = 4
cnt = 10**4

prNums = lcg(seed, cnt, a, c, m)
plot(prNums)
show(prNums, True)

```
